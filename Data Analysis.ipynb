###### Pivot table

```python
mcd1 = pd.read_csv("./Mcdonalds.csv")
```

```python
pt = mcd1.pivot_table(index ='Item',columns='Category',values='Calories',aggfunc=[min,max])
```

```python
pt
```

```python
pt.reset_index(inplace=True)
```

```python
pt.head()
```

```python
pt.fillna(0, inplace=True)
pt.head()
```

```python
pt.melt(id_vars='Item', value_name='Calories_new',var_name='Category_min/max').head()
```
